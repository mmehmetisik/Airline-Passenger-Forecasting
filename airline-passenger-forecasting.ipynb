<h3 style="text-align: center; color: #d35400; font-size: 1.8em; border-bottom: 2px solid #e74c3c; padding-bottom: 5px; margin-bottom: 20px;">üìâ Understanding the Smoothing Method</h3>

<p style="font-size: 1.2em; color: #7f8c8d; margin-top: 20px;">The concept of the smoothing method revolves around refining time series data for better analysis and forecasting. Its core function is to minimize the data noise, emphasizing the principal trend or pattern, which is crucial for deriving insightful and accurate predictions.</p>

<h4 style="color: #2c3e50; font-size: 1.6em; margin-top: 30px; border-left: 5px solid #e74c3c; padding-left: 15px;">üìò Categories of Smoothing Methods</h4>

<ul style="list-style-type: none; padding: 0; margin-top: 15px; font-size: 1.1em;">
    <li style="padding: 8px 0; color: #8e44ad;"><strong>Moving Average Approach:</strong> In this method, a specific window width is selected. The average of the data points within this window forms a new data point. The window then slides over the dataset, calculating moving averages throughout.</li>
    <li style="padding: 8px 0; color: #8e44ad;"><strong>Technique of Exponential Smoothing:</strong> Predictions are created by obtaining the weighted average of previous data points. Here, the latest data is assigned more significance, while older data is given less. It operates on the premise that data has a stronger connection with its recent past.</li>
    <li style="padding: 8px 0; color: #8e44ad;"><strong>Exponential Smoothing Subcategories:</strong> The technique of exponential smoothing further branches out into Single, Double, and Triple Exponential Smoothing. They serve to model different components like level, trend, and seasonality present in the dataset.</li>
</ul>

<p style="font-size: 1.2em; color: #7f8c8d; margin-top: 20px; border-top: 2px dashed #95a5a6; padding-top: 20px;">üåç In domains like stock trading, power usage, and meteorological predictions, smoothing methods are widely adopted.</p>

<div style="background-color: #e8f1f5; padding: 40px; border-radius: 25px; box-shadow: 0 0 20px rgba(0,0,0,0.07); font-family: 'Verdana', sans-serif; line-height: 1.7; border: 3px solid #bed2d7;">

<h2 style="text-align: center; color: #1a434f; font-size: 2.1em; border-bottom: 3px dotted #3498db; padding-bottom: 15px; margin-bottom: 35px;">üìä Diving into Exponential Smoothing</h2>

<h3 style="color: #2c8e99; font-size: 1.85em; margin-top: 30px;">üìò Single Exponential Smoothing (SES)</h3>
<ul style="font-size: 1.2em; color: #2e4057; margin-top: 25px;">
    <li>Best suited for stationary series without trends or seasonality.</li>
    <li>It primarily models the 'Level' component, akin to the series average.</li>
    <li>Predictions lean more towards recent past data, using it as the weighted foundation.</li>
    <li>An ideal method for trend-less, non-seasonal series.</li>
</ul>

<h3 style="color: #2c8e99; font-size: 1.85em; margin-top: 35px;">üìó Double Exponential Smoothing (DES)</h3>
<ul style="font-size: 1.2em; color: #2e4057; margin-top: 25px;">
    <li>Capable of modeling both 'Level' and 'Trend' components, excluding seasonality.</li>
    <li>An evolution to SES, encompassing the trending aspect of series.</li>
</ul>

<h3 style="color: #2c8e99; font-size: 1.85em; margin-top: 35px;">üìô Triple Exponential Smoothing (TES - Holt-Winters)</h3>
<ul style="font-size: 1.2em; color: #2e4057; margin-top: 25px;">
    <li>The pinnacle of exponential smoothing techniques.</li>
    <li>It holistically models 'Level', 'Trend', and 'Seasonality' components.</li>
    <li>Adaptable to both stationary series and those exhibiting trends or seasonality.</li>
</ul>

<h4 style="text-align: center; color: #e67e22; font-size: 1.7em; margin-top: 45px; border-top: 2px solid #b4c0c7; padding-top: 20px;">üîç Picking the Right Method</h4>
<p style="font-size: 1.2em; color: #2e4057; margin-top: 25px;">The key is to opt for the method showcasing the most promising results for the dataset, gauged by its performance metrics.</p>

<div style="background-color: #f4f4f4; padding: 20px; border-radius: 20px; box-shadow: 0 0 25px rgba(0,0,0,0.1); font-family: Arial, sans-serif; line-height: 1.6; border: 5px solid #3498db;">
<h2 style="text-align: center; color: #2c3e50; font-size: 2em; border-bottom: 3px double #2980b9; padding-bottom: 15px; margin-bottom: 30px;">üìà Components of a Time Series</h2>
</div>

<div style="background-color: #f4f4f4; padding: 20px; border-radius: 20px; box-shadow: 0 0 25px rgba(0,0,0,0.1); font-family: Arial, sans-serif; line-height: 1.6; border: 5px solid #3498db;">
<h3 style="color: #3498db; font-size: 1.8em; margin-top: 30px;">Trend:</h3>
<p>This represents the long-term movement of the time series. For example, when you review the sales of a company over the years, you might observe a trend of increasing or decreasing sales. The trend typically changes in a linear or non-linear manner over time.</p>
</div>

<div style="background-color: #f4f4f4; padding: 20px; border-radius: 20px; box-shadow: 0 0 25px rgba(0,0,0,0.1); font-family: Arial, sans-serif; line-height: 1.6; border: 5px solid #3498db;">
<h3 style="color: #3498db; font-size: 1.8em; margin-top: 30px;">Seasonality:</h3>
<p>This refers to short-term fluctuations in the time series that repeat at specific intervals. For instance, during certain months of the year (e.g., summer vacation or winter season), a tourism company might observe spikes in their revenue. Seasonality is generally a component with a known and fixed period.</p>
</div>

<div style="background-color: #f4f4f4; padding: 20px; border-radius: 20px; box-shadow: 0 0 25px rgba(0,0,0,0.1); font-family: Arial, sans-serif; line-height: 1.6; border: 5px solid #3498db;">
<h3 style="color: #3498db; font-size: 1.8em; margin-top: 30px;">Cycle:</h3>
<p>This denotes long-term oscillations and is usually associated with economic cycles. Cycles differ from seasonality since they typically do not have a known fixed period. For example, fluctuations might be observed between periods of economic expansion and contraction.</p>
</div>

<div style="background-color: #f4f4f4; padding: 20px; border-radius: 20px; box-shadow: 0 0 25px rgba(0,0,0,0.1); font-family: Arial, sans-serif; line-height: 1.6; border: 5px solid #3498db;">
<h3 style="color: #3498db; font-size: 1.8em; margin-top: 30px;">Randomness (Error or Noise):</h3>
<p>This component represents random fluctuations that are not explained by the other three components. Randomness in the time series is unforeseen oscillations and is often termed "noise." Randomness is generally a component challenging to model or predict since it results from unforeseen events.</p>
</div>

<div style="background-color: #f4f4f4; padding: 20px; border-radius: 20px; box-shadow: 0 0 25px rgba(0,0,0,0.1); font-family: Arial, sans-serif; line-height: 1.6; border: 5px solid #3498db

<h2 style="text-align: center; color: #1e272e; font-size: 2.1em; border-bottom: 3px dashed #3498db; padding-bottom: 12px; margin-bottom: 35px;">üìä Dive into Time Series: Statistical Methods</h2>

<section style="background-color: #d6eaf8; padding: 25px; border-radius: 20px; margin-bottom: 20px;">
    <h3 style="color: #c0392b; font-size: 1.75em; margin-top: 0; font-weight: 500;">AR (Auto-Regressive Model)</h3>
    <p>Characterized by its reliance on past values, the AR model is tailored for stationary series without trends or seasonality.</p>
    <p>The model equation:</p>
    <p>\[ X_t = c + \phi_1 X_{t-1} + \phi_2 X_{t-2} + \dots + \phi_p X_{t-p} + \varepsilon_t \]</p>
</section>

<section style="background-color: #d6eaf8; padding: 25px; border-radius: 20px; margin-bottom: 20px;">
    <h3 style="color: #d35400; font-size: 1.75em; margin-top: 0; font-weight: 500;">MA (Moving Average Model)</h3>
    <p>The MA model is influenced by past white noise or random shocks, offering a unique perspective on time series analysis.</p>
    <p>The model equation:</p>
    <p>\[ X_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \dots + \theta_q \varepsilon_{t-q} \]</p>
</section>

<section style="background-color: #d6eaf8; padding: 25px; border-radius: 20px; margin-bottom: 20px;">
    <h3 style="color: #8e44ad; font-size: 1.75em; margin-top: 0; font-weight: 500;">ARMA (Auto-Regressive Moving Average Model)</h3>
    <p>Blending the essence of both AR and MA, the ARMA model introduces a versatile approach to time series forecasting.</p>
    <p>The model equation:</p>
    <p>\[ X_t = c + \phi_1 X_{t-1} + \dots + \phi_p X_{t-p} + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \dots + \theta_q \varepsilon_{t-q} \]</p>
</section>

<section style="background-color: #d6eaf8; padding: 25px; border-radius: 20px; margin-bottom: 20px;">
    <h3 style="color: #27ae60; font-size: 1.75em; margin-top: 0; font-weight: 500;">ARIMA (Auto-Regressive Integrated Moving Average Model)</h3>
    <p>ARIMA is a refined variant of ARMA, integrating differencing to achieve stationarity in the time series data.</p>
    <p>ARIMA is denoted as ARIMA(p, d, q), where:</p>
    <ul>
        <li>\( p \) represents the lags for the AR model.</li>
        <li>\( d \) indicates the differencing steps.</li>
        <li>\( q \) specifies the lags for the MA model.</li>
    </ul>
</section>

<section style="background-color: #d6eaf8; padding: 25px; border-radius: 20px;">
    <h3 style="color: #2980b9; font-size: 1.75em; margin-top: 0; font-weight: 500;">SARIMA (Seasonal Auto-Regressive Integrated Moving Average Model)</h3>
    <p>Expanding on ARIMA, SARIMA focuses on seasonal patterns in data.</p>
    <p>SARIMA is represented as SARIMA(p, d, q)(P, D, Q)[s], where:</p>
    <ul>
        <li>\( p, d, q \) are the parameters for ARIMA.</li>
        <li>\( P, D, Q \) denote the seasonal components.</li>
        <li>\( s \) is the seasonal period.</li>
    </ul>
</section>

</div>

# <div style="padding: 30px; color:white; margin:10; font-size:75%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>6 |</span></b> <b>Single Exponential Smoothing</b></div>

```python
def ses_optimizer(train, alphas, step=48):
    best_alpha, best_mae = None, float("inf")
    for alpha in alphas:
        ses_model = SimpleExpSmoothing(train).fit(smoothing_level=alpha)
        y_pred = ses_model.forecast(step)
        mae = mean_absolute_error(test, y_pred)
        if mae < best_mae:
            best_alpha, best_mae = alpha, mae
        print("alpha:", round(alpha, 2), "mae:", round(mae, 4))
    print("best_alpha:", round(best_alpha, 2), "best_mae:", round(best_mae, 4))
    return best_alpha, best_mae
```

```python
alphas = np.arange(0.01, 1, 0.10)
```

```python
best_alpha, best_mae = ses_optimizer(train, alphas, step=24)
```

```python
ses_model = SimpleExpSmoothing(train).fit(smoothing_level=best_alpha)
```

```python
y_pred = ses_model.forecast(24)
```

```python
y_pred
```

```python
def plot_prediction(y_pred, label):
    train["total_passengers"].plot(legend=True, label="TRAIN")
    test["total_passengers"].plot(legend=True, label="TEST")
    y_pred.plot(legend=True, label="PREDICTION")
    plt.title("Train, Test and Predicted Test Using "+label)
    plt.show(block=True)
```

```python
plot_prediction(y_pred, "Single Exponential Smoothing")
```

# <div style="padding: 30px; color:white; margin:10; font-size:75%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>7 |</span></b> <b>Double Exponential Smoothing</b></div>

```python
def des_optimizer(train, alphas, betas, step=48):
    best_alpha, best_beta, best_mae = None, None, float("inf")
    for alpha in alphas:
        for beta in betas:
            des_model = ExponentialSmoothing(train, trend="add").fit(smoothing_level=alpha, smoothing_slope=beta)
            y_pred = des_model.forecast(step)
            mae = mean_absolute_error(test, y_pred)
            if mae < best_mae:
                best_alpha, best_beta, best_mae = alpha, beta, mae
            print("alpha:", round(alpha, 2), "beta:", round(beta, 2), "mae:", round(mae, 4))
    print("best_alpha:", round(best_alpha, 2), "best_beta:", round(best_beta, 2), "best_mae:", round(best_mae, 4))
    return best_alpha, best_beta, best_mae
```

```python
alphas = np.arange(0.01, 1, 0.10)
```

```python
betas = np.arange(0.01, 1, 0.10)
```

```python
best_alpha, best_beta, best_mae = des_optimizer(train, alphas, betas, step=24)
```

```python
des_model = ExponentialSmoothing(train, trend="add").fit(smoothing_level=best_alpha,
                                                         smoothing_slope=best_beta)
```

```python
y_pred = des_model.forecast(24)
```

```python
y_pred
```

```python
plot_prediction(y_pred, "Double Exponential Smoothing")
```

# <div style="padding: 30px; color:white; margin:10; font-size:75%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>8 |</span></b> <b>Triple Exponential Smoothing (Holt-Winters)</b></div>

```python
def tes_optimizer(train, abg, step=48):
    best_alpha, best_beta, best_gamma, best_mae = None, None, None, float("inf")
    for comb in abg:
        tes_model = ExponentialSmoothing(train, trend="add", seasonal="add", seasonal_periods=12).\
            fit(smoothing_level=comb[0], smoothing_slope=comb[1], smoothing_seasonal=comb[2])
        y_pred = tes_model.forecast(step)
        mae = mean_absolute_error(test, y_pred)
        if mae < best_mae:
            best_alpha, best_beta, best_gamma, best_mae = comb[0], comb[1], comb[2], mae
        print([round(comb[0], 2), round(comb[1], 2), round(comb[2], 2), round(mae, 2)])

    print("best_alpha:", round(best_alpha, 2), "best_beta:", round(best_beta, 2), "best_gamma:", round(best_gamma, 2),
          "best_mae:", round(best_mae, 4))

    return best_alpha, best_beta, best_gamma, best_mae
```

```python
alphas = betas = gammas = np.arange(0.10, 1, 0.20)
```

```python
abg = list(itertools.product(alphas, betas, gammas))
```

```python
best_alpha, best_beta, best_gamma, best_mae = tes_optimizer(train, abg, step=24)
```

```python
tes_model = ExponentialSmoothing(train, trend="mul", seasonal="mul", seasonal_periods=12).\
            fit(smoothing_level=best_alpha, smoothing_slope=best_beta, smoothing_seasonal=best_gamma)
```

```python
y_pred = tes_model.forecast(24)
```

```python
plot_prediction(y_pred, "Triple Exponential Smoothing ADD")
```

# <div style="padding: 30px; color:white; margin:10; font-size:75%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>9 |</span></b> <b>ARIMA(p, d, q): (Autoregressive Integrated Moving Average)</b></div>

```python
p = d = q = range(0, 4)
```

```python
pdq = list(itertools.product(p, d, q))
```

```python
def arima_optimizer_aic(train, orders):
    best_aic, best_params = float("inf"), None
    for order in orders:
        try:
            arma_model_result = ARIMA(train, order).fit(disp=0)
            aic = arma_model_result.aic
            if aic < best_aic:
                best_aic, best_params = aic, order
            print('ARIMA%s AIC=%.2f' % (order, aic))
        except:
            continue
    print('Best ARIMA%s AIC=%.2f' % (best_params, best_aic))
    return best_params
```

```python
best_params_aic = arima_optimizer_aic(train, pdq)
```

```python
arima_model = ARIMA(train, best_params_aic).fit()
```

```python
y_pred = arima_model.forecast(24)
```

```python
y_pred
```

```python
mean_absolute_error(test, y_pred)
```

```python
plot_prediction(pd.Series(y_pred, index=test.index), "ARIMA")
```
